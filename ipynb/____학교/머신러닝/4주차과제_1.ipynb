#### 모듈불러오기

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

#### 함수정의

```python
def f1(x0,x1):
    r = x0**2 + x1**2
    ans = r *np.exp(-r)
    return ans

def f2(x0,x1):
    r = x0**2 + x1**2
    ans = r *np.exp(-r)
    return ans * 0.3
```

#### 연산

```python
n = 40
x = np.linspace(-3,3,n) #n 등분
y = np.linspace(-3,3,n)
x, y = np.meshgrid(x, y)
z1  = f1(x, y)
z2  = f2(x, y)
```

#### 설정

```python
fig = plt.figure(figsize=(18, 9)) # 출력화면크기

# 첫번째 그래프
ax1 = fig.add_subplot(1,2,1, projection='3d') # 세번째인자는 n번째 plot
ax1.plot_surface(x, y, z1, rstride=1, cstride=1, alpha=0.3, color='green', edgecolor='black')
ax1.set_zticks(np.arange(0, 0.5, 0.1))
ax1.view_init(30, -45) # 보는 각도
plt.xlabel('x')
plt.ylabel('y')

# 두번째 그래프
ax2 = fig.add_subplot(1,2,2, projection='3d') # 세번째인자는 n번째 plot
ax2.plot_surface(x, y, z2, rstride=1, cstride=1, alpha=0.3, color='green', edgecolor='black')
ax2.set_zticks(np.arange(0, 0.5, 0.1))
ax2.view_init(30, -45) # 보는 각도
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```
