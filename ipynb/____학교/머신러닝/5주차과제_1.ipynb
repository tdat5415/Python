### 2차원 그래프 그리기

```python
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib inline

np.random.seed(1) # 시드 값에 따라 난수와 흡사하지만 항상 같은 결과를 반환
x = np.arange(10)
y = np.random.rand(10) # 표준정규분포

print("\t x = ",x)
print("\t y = ",np.round(y,2))

plt.plot(x,y)
plt.show()
##################################################
def f(x):
    return (x-2) * x * (x+2) # x^3 - 4x

print(" f(1) = ",f(1))
print(f(np.array([1,2,3])))

x = np.arange(-3,3.5,.5) # 0.5씩 끊기
print(' x = ',x)

x = np.linspace(-3,3,13) # 13등분하기
print(' x = ',np.round(x,1)) # 소수점1자리까지반올림
plt.plot(x,f(x))
plt.show()
#################################################
def f2(x,w):
    return (x-w) * x * (x+2)

x= np.linspace(-3,3,100) # 범위안 100등분하기
plt.plot(x,f2(x,2),color='black',label = '$w=2$')
plt.plot(x,f2(x,1),color='cornflowerblue',label = '$w=1$')
plt.legend(loc="upper left") # 그래프이름 출력위치
plt.ylim(-15,15) # 범위제한
plt.title('$f_2(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid(True)
plt.show()
```

### 여러개의 그래프 그리기

```python
import numpy as np
import matplotlib.pyplot as plt

def f2(x,w):
    return (x-w) * x * (x+2)

x= np.linspace(-3,3,100) # 범위안 100등분하기

plt.figure(figsize=(10,3)) # 화면크기
plt.subplots_adjust(wspace=0.5, hspace=0.5) # 사이간격 조절
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.plot(x,f2(x,i),color='black')
    plt.title(i+1)
    plt.ylim(-20,20) # 범위제한
    plt.grid(True)
plt.show()
```

### 3차원 그래프 그리기(이변수 함수, (수치를 색으로 표시하기)

```python
import numpy as np
import matplotlib.pyplot as plt

def f3(x0,x1):
    r = 2*x0**2 + x1**2
    ans = r *np.exp(-r)
    return ans

xn = 9
x0 = np.linspace(-2,2,xn)
x1 = np.linspace(-2,2,xn)
y  = np.zeros((len(x0),len(x1)))
print(" x0 = ",x0)
print(" x1 = ",x1)

for i0 in range(xn):
    for i1 in range(xn):
        y[i1,i0] = f3(x0[i0],x1[i1])

print(' y = \n',np.round(y,1))

plt.figure(figsize=(3.5,3)) # 화면크기
plt.xlabel('$x0$')
plt.ylabel('$x1$')
# plt.jet()
plt.pcolor(y, cmap=plt.cm.hot) # 2차원 배열값을 색으로 
plt.colorbar() # 컬러바 출력
plt.show()
```

### 3차원 그래프 그리기_함수의 표면을 표시하기(surface)

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def f3(x0,x1):
    r = 2*x0**2 + x1**2
    ans = r *np.exp(-r)
    return ans

xn = 40
x0 = np.linspace(-2,2,xn)
x1 = np.linspace(-3,3,xn)
y  = np.zeros((len(x0),len(x1)))

for i0 in range(xn):
    for i1 in range(xn):
        y[i1,i0] = f3(x0[i0],x1[i1])

print(' x0 = ')
print(np.round(x0,1))
print(' x1 = ')
print(np.round(x1,1))

xx0, xx1 = np.meshgrid(x0, x1)
print(' xx0 = ')
print(np.round(xx0,1))
print(' xx1 = ')
print(np.round(xx1,1))


plt.figure(figsize=(8, 5))
ax = plt.subplot(1,1,1, projection='3d') # 3차원 그래프
ax.plot_surface(xx0, xx1, y, rstride=1, cstride=1, \
                alpha=0.3, color='pink', edgecolor='black') # 세부설정
ax.set_zticks((0, 0.2))
ax.view_init(45, -55) # 보는 각도
plt.xlabel('$x_0$')
plt.ylabel('$x_1$')
plt.show()
```

### 3차원 그래프 그리기_등고선으로 표시하기(surface)

```python
import numpy as np
import matplotlib.pyplot as plt
# from mpl_toolkits.mplot3d import Axes3D

def f3(x0,x1):
    r = 2*x0**2 + x1**2
    ans = r *np.exp(-r)
    return ans

xn = 50
x0 = np.linspace(-2,2,xn) # 범위안 n등분하기
x1 = np.linspace(-2,2,xn) # 범위안 n등분하기
y = np.zeros((len(x0),len(x1)))

for i0 in range(xn):
    for i1 in range(xn):
        y[i1,i0] = f3(x0[i0],x1[i1])

xx0,xx1 = np.meshgrid(x0,x1)

plt.figure(1,figsize=(5,4)) # 화면크기

plt.contourf(xx0,xx1,y,5,cmap=plt.cm.hot) # 색설정
cont = plt.contour(xx0,xx1,y,5,colors='black') # 등고선
cont.clabel(fmt='%3.2f',fontsize=8)
plt.xlabel('$x_0$',fontsize=14)
plt.ylabel('$x_1$',fontsize=14)
plt.show()
```
