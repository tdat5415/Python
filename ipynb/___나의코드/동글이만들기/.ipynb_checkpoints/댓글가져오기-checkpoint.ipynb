## 모듈 불러오기

```python
import re
import numpy as np
import pyperclip
import pyautogui as pag
from time import sleep
import pandas as pd
import copy
```

## 전처리함수

```python
def pre_pro(M):
    words = []
    for m in M: # '와 ㅋㅋㅋㅋ'
        for w in m.split(): # ['와', 'ㅋㅋㅋㅋ']
            if '<' in w:
                ind = w.find('<')
                w = w[:ind]
            if ';' in w or w == '':
                continue
            if w == w[::-1]: #
                words.append(w[:2])
                continue
            words.append(w)
    
    return words
```

## 문장 모으기

```python
sleep(1)
for i in range(10):
    pag.click(x=1545, y=207, button='right')
    pag.click(x=1598, y=303)
    pag.click(x=1769, y=330)
    sleep(0.1)
    raw_code = pyperclip.paste()
    
    pattern = r'<span class="text-fragment" data-a-target="chat-message-text">(.{1,20})</span>' # .는 모든문자하나 {1,20}은 길이범위
    p = re.compile(pattern)
    msgs = p.findall(raw_code)
    all_msg.extend(msgs)
    
    print(i, len(all_msg))
    
    for j in range(60):
        sleep(1)

print(len(all_msg))
```

## 문장을 단어로

```python
### 문장을 단어로
all_words = pre_pro(all_msg)

### 단어 카운트
word_dict = {}
for word in all_words:
    try:
        word_dict[word] += 1
    except:
        word_dict[word] = 1
        
### 정렬후 리스트화
sorted_words = list(sorted(word_dict.items(), key = lambda x : x[1], reverse=True))
for i in range(len(sorted_words)):
    sorted_words[i] = list(sorted_words[i])

copy_words = copy.deepcopy(sorted_words)
### 추가 처리
for i in range(len(copy_words)):
    for j in range(len(copy_words)):
        if j == i:
            continue
        if len(copy_words[i][0]) <= 3:
            break
        if copy_words[i][0] in copy_words[j][0]:
            sorted_words[i][1] += copy_words[j][1]

### 다시 정렬
sorted_words = sorted(sorted_words, key = lambda x : x[1], reverse=True)
            
sorted_words[:40]
```
